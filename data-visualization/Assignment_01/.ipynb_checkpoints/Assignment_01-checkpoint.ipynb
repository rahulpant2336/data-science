#### 1: View and add the dataset.

```python
#Import the required library
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
#View the plot in notebook
%matplotlib inline
```

```python
#Import the dataset
auto_mpg_data = pd.read_csv('auto_data.csv')
```

```python
#View the top 5 reords
auto_mpg_data.head()
```

#### 2: Write a user-defined function for origin  


```python
#use apply function
# 1-USA, 2-Europe, 3-Asia
def my_origin(val):
    if val==1:
        return 'USA'
    elif val==2:
        return 'Europe'
    else:
        return 'Asia'

auto_mpg_data['origin'] = auto_mpg_data['origin'].apply(my_origin)

    
```

```python
#view first 30 data points after applying the user defined function to data set
auto_mpg_data.head(30)
```

#### 3: Draw the pair plot using sns for mpg, weight,orign and with hue origin, set the size to 4


```python
sns.pairplot(auto_mpg_data[['mpg','weight','origin']],hue='origin',size=4)
```
