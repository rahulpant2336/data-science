#### 6: View the article names and navigate through them

```python
#Create article object for finding the list (Hint: the class information is obtained from the web page (view page source))
article_name = sl_soup.find(class_='testimonial_div_inner')
```

```python
#View the type of the article object
type(article_name)
```

```python
#Find all the article names
article_name.findAll('p')
```

```python
#Find first article name
find_first_article = article_name.li
print(find_first_article.p)
```

```python
#Find next sibling (next article name)
find_next_article = find_first_article.next_sibling.next_sibling
print(find_next_article.p)
```

```python
#Find previous sibling (previous article name)
find_previous_article = find_next_article.previous_sibling.previous_sibling
print(find_previous_article.p)
```
